```python
import pyterrier as pt
if not pt.started():
  pt.init()
```

```python
indexer = pt.TRECCollectionIndexer("/content/drive/MyDrive/Data/polygon_index2",
                          meta={"docno":300,"title":1024},meta_tags={"title":"title"})
indexref = indexer.index("/content/drive/MyDrive/Data/polygon.trec")
index = pt.IndexFactory.of(indexref)
print(index.getCollectionStatistics().toString())
```

```python

tf_idf = pt.BatchRetrieve(index, wmodel="TF_IDF")
bm25 = pt.BatchRetrieve(index, wmodel="BM25")
lm = pt.BatchRetrieve(index, wmodel="Hiemstra_LM")
```

```python
import pandas as pd
queries = pd.DataFrame([["q1","Pokemon"],["q2","Dragon"]],columns=["qid","query"])
```

```python
queries
```

```python
tf_idf.transform(queries)
```

```python
qrels = pd.DataFrame([["q1","NYT20000131.0026",1],
                      ["q1","NYT20000130.0020",1],
                      ["q2","NYT20000131.0308",1],
                      ["q2","NYT20000131.0139",1]], columns=["qid","docno","label"])
```

```python
queries
```

```python
qrels
```

```python
pt.Experiment (
    [tf_idf,bm25,lm],
    queries,
    qrels,
    eval_metrics=["map","Rprec","recip_rank"]
)
```

```python
dataset = pt.get_dataset("vaswani")
indexer = pt.TRECCollectionIndexer("./vsawani")
indexref = indexer.index(dataset.get_corpus())
index = pt.IndexFactory.of(indexref)
print(index.getCollectionStatistics().toString())
```

```python
tf_idf = pt.BatchRetrieve(index, wmodel="TF_IDF")
bm25 = pt.BatchRetrieve(index, wmodel="BM25")
lm = pt.BatchRetrieve(index, wmodel="Hiemstra_LM")
```

```python
queries = pd.DataFrame([["4","systems of data coding for information transfer"],
 ["6","number representation in binary machines"]],columns=["qid","query"])
```

```python
pt.Experiment (
    [tf_idf,bm25,lm],
    queries,
    dataset.get_qrels(),
    eval_metrics=["map","Rprec","recip_rank"]
)
```
